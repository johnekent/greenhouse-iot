```python
df = pd.read_csv("c:\\temp\\Book1.csv")
df
```

```python
import re
def find_match(str):
    match = re.search('[ralsthng]', str)
    if match:
        return True
    else:
        return False
```

```python
find_match('apple')
```

```python
find_match('XXIIXX')
```

```python
df['match'] = df['Unnamed: 0'].apply(find_match)
```
